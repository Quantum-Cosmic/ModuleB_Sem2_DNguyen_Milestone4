# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

**Advantages and disadvantages:**

| Aspect | Graph 1: Separate Plots | Graph 2: Overlay Lines |
|--------|-------------------------|-------------------------|
| Comparison | Difficult. Eyes must jump between plots | Easy. Direct visual comparison |
| Pattern | Harder to spot differences in trends | Clearer to see if patterns change |
| Clarity | Each week's pattern is isolated and clear | Lines can cross and become messy |
| Visual | Clean, no overlapping lines | Can get cluttered |
| Scale | Requires y-axis alignment | Automatic same scale |

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
# For this week, I choose Figure 8.5 (Page 193)

# Create the data for each product
years = ['2019', '2020', '2021', '2022', '2023', '2024', '2025']
year_positions = [0, 1, 2, 3, 4, 5, 6]

# Product data (only available years)
product_data = {
    'Product A': [395, 410, 420, 380, 300, 280, 270],
    'Product B': [370, 400, 410, 370, 290, 270, 260],
    'Product C': [None, None, 100, 180, 200, 250, 180],
    'Product D': [None, None, None, 165, 275, 220, 210],
    'Product E': [None, None, None, None, None, 100, 205]
}

# Create figure
fig, ax = plt.subplots(figsize=(10, 6))

# Color palette for the lines
colors = ['#1f77b4', '#d62728', "#63c263", "#5b0c6f", "#178de2"]

# Plot each product
for i, (product, values) in enumerate(product_data.items()):
    # Filter out None values for plotting
    valid_years = [year_positions[j] for j, v in enumerate(values) if v is not None]
    valid_values = [v for v in values if v is not None]
    
    linewidth = 3.5
    color = colors[i]
    
    ax.plot(valid_years, valid_values, color=color, linewidth=linewidth, label=product)

# Customize the chart
ax.set_xlim(-0.5, 6.5)
ax.set_ylim(0, 500)

# Set x-axis ticks
ax.set_xticks(year_positions)
ax.set_xticklabels(years, fontsize=10)

# Add title
ax.set_title('Average Retail Product Price Per Year\n', fontsize=14, fontweight='bold', loc='center')

# Add light grid on y-axis only
ax.grid(axis='y', alpha=0.7, linestyle='--', linewidth=0.5)

# Set y-axis with ticks
ax.set_yticks([0, 100, 200, 300, 400, 500])
ax.set_yticklabels(['$0', '$100', '$200', '$300', '$400', '$500'], fontsize=10)

# Clean up spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)

# Legend at bottom center
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.08), 
          ncol=7, frameon=False, fontsize=10)

plt.tight_layout()
plt.show()
```
