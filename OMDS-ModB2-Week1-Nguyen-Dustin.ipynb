# 1. Lesson on how to search for Python commands

Let's consider a few possible ways to learn about Python programming.  Let's suppose you want to learn how to produce a short summary of the information in your DataFrame.

1. Your **instructor** could provide the information.

You could be provided with a lesson about functions like info() and describe().  If you have a pandas DataFrame called df, then you can summarize its contents using df.info() or df.describe().  df.info() provides a list of column names with their counts and data types.  df.describe() will provide information such as the mean, min, max, standard deviation, and quantiles.  Thus:

```python
df = pd.DataFrame([[1, 4], [2, 5], [3, 6], [4, 7]], columns = ['A', 'B'])
df.describe()
```

In this describe() result, we see that the two columns A and B each have four elements.  The means and other statistics are shown.

2. You could look up the information on **Google**.

If I Google the question "how do I briefly summarize the contents of a dataframe using Python," I receive the following link (among others), which discusses the describe() command mentioned above:

https://www.w3schools.com/python/pandas/ref_df_describe.asp

It also provide the complete usage information:

dataframe.describe(percentiles, include, exclude, datetime_is_numeric)

It explains that "percentiles" is set by default to [0.25, 0.5, 0.75] but we could change that.  Let's try it!  Since there are three intervals here rather than four, it might be more meaningful to ask about a 33rd and 67th percentile rather than 25, 50, and 75.  We can use 1/3 for 0.33 and 2/3 for 0.67 to get the exact percentile values.

```python
df = pd.DataFrame([[1, 4], [2, 5], [3, 6], [4, 7]], columns = ['A', 'B'])
df.describe(percentiles = [1/3, 2/3])
```

Apparently, the 50% value (the median) stays even though we did not specifically request it.

```python
# I'll recreate the plot to show better visually the data points 
# that are above 2026, which are data errors. 
# I'll color those red and the rest blue, 
# and add a horizontal line at 2026 to show the cutoff.

# Prepare the data
df_bank_clean['BirthDate'] = pd.to_datetime(df_bank_clean['BirthDate'])
df_bank_clean['AccountOpened'] = pd.to_datetime(df_bank_clean['AccountOpened'])

# Create masks for data above and below 2026
above_2026 = df_bank_clean['AccountOpened'].dt.year > 2026
below_2026 = df_bank_clean['AccountOpened'].dt.year <= 2026

# Create the plot
ax = plt.gca()

# Plot points below or at 2026 in blue
plt.scatter(df_bank_clean.loc[below_2026, 'BirthDate'].dt.year, 
           df_bank_clean.loc[below_2026, 'AccountOpened'].dt.year, 
           c='blue', label='Valid dates (<= 2026)')

# Plot points above 2026 in red
plt.scatter(df_bank_clean.loc[above_2026, 'BirthDate'].dt.year, 
           df_bank_clean.loc[above_2026, 'AccountOpened'].dt.year, 
           c='red', label='Future dates (> 2026)')

# Add horizontal line at 2026
plt.axhline(y=2026, color='green', linestyle='--', linewidth=2, label='Current Year (2026)')

# Add labels and title
plt.xlabel('Birth Date', fontsize=12)
plt.ylabel('Account Opened', fontsize=12)
plt.title('Birth Year vs Account Opening Year\n', fontsize=14)

# Add legend
plt.legend()

# Show the plot
plt.tight_layout()
```

```python
# I'll use pie chart to show the distribution of account types for valid data vs future dates.
# I've got warning about using .loc to set values on a copy of a slice from a DataFrame, so I'll use .loc to avoid that warning.
# Looked up online and changed the code for the "Prepare the data" section

# Prepare the data
df_bank_clean.loc[:, 'BirthDate'] = pd.to_datetime(df_bank_clean['BirthDate'])
df_bank_clean.loc[:, 'AccountOpened'] = pd.to_datetime(df_bank_clean['AccountOpened'])
df_bank_clean.loc[:, 'OpenYear'] = df_bank_clean['AccountOpened'].dt.year

# Split data into valid (≤2026) and invalid (>2026)
valid_data = df_bank_clean[df_bank_clean['OpenYear'] <= 2026]
invalid_data = df_bank_clean[df_bank_clean['OpenYear'] > 2026]

# Get account type counts for each group
valid_counts = valid_data['AccountType'].value_counts()
invalid_counts = invalid_data['AccountType'].value_counts()

# Create pie charts
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))

# Colors for account types
colors = {'checking': 'pink', 'savings': 'yellow', 'cd': 'lightblue'}

# Pie chart 1: Valid data (≤2026)
wedges1, texts1, autotexts1 = ax1.pie(valid_counts.values, 
                                       labels=valid_counts.index, 
                                       colors=[colors[acct] for acct in valid_counts.index],
                                       autopct='%1.1f%%',
                                       startangle=90,
                                       explode=[0.05] * len(valid_counts))
ax1.set_title(f'Account Type\nValid Data (≤2026)\nTotal: {len(valid_data)} accounts', fontsize=12)

# Pie chart 2: Invalid data (>2026)
wedges2, texts2, autotexts2 = ax2.pie(invalid_counts.values, 
                                       labels=invalid_counts.index, 
                                       colors=[colors[acct] for acct in invalid_counts.index],
                                       autopct='%1.1f%%',
                                       startangle=90,
                                       explode=[0.05] * len(invalid_counts))
ax2.set_title(f'Account Type\nFuture Dates (>2026)\nTotal: {len(invalid_data)} accounts', fontsize=12)

# Add a main title
fig.suptitle('Account Type Distribution: Valid Data vs Future Dates\n', fontsize=14, fontweight='bold')

plt.tight_layout()
```
