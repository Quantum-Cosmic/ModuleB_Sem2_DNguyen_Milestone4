# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series_2 =  100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series_2)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

```python
# Re-plot last three months with multiple preattentive attributes:
# 1. Larger size (for better visual)
# 2. Bold color (red), (gray) for other months
# 3. Different marker (star to hi-light)
# 4. Thicker line (emphasize)
# 5. Added data labels (show numbers on top)


# 1. Setup a plot with larger size
plt.figure(figsize=(8, 5))

# 2. Plot all data in gray color (dimmer)
plt.plot(datearray, time_series_2, 'o-', color='lightgray', linewidth=1, markersize=3, alpha=0.7)

# 2. Bold color (red), (gray) for other months
# 3. Different marker (star to hi-light)
# 4. Thicker line (emphasize)
plt.plot(datearray[-3:], time_series_2[-3:], '*-', color='red', linewidth=4, markersize=14, 
         markerfacecolor='yellow', markeredgecolor='darkred')

# 5. Added data labels (show numbers on top)
for i in range(-3, 0):
    plt.text(datearray[i], time_series_2[i] + 8, f'{time_series_2[i]:.0f}', 
             ha='center', fontsize=11, fontweight='bold', color='red')

# Add a highlight box to get attention
plt.axvspan(datearray[-3], datearray[-1], alpha=0.1, color='red')

# Change title to get attention
plt.title('CRITICAL: Last Three Months Show Sharp Decline\n', fontsize=14, fontweight='bold', color='darkred')

plt.xlabel('\nYear and Month', fontsize=12)
plt.ylabel('Phosphate Level (% of eutrophic value)', fontsize=12)
plt.grid(True, alpha=0.2)
plt.tight_layout()
plt.show()
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
# This takes a whole lot of time to customize, Why not use Excel for this kind of charts?

#  Create Priority names list
priorities = [
    'Education',
    'Agriculture & rural development',
    'Poverty reduction',
    'Reconstruction',
    'Economic growth',
    'Health',
    'Job creation',
    'Governance',
    'Anti-corruption',
    'Transport',
    'Energy',
    'Law & Justice',
    'Basic infrastructure',
    'Public sector reform',
    'Public financial management'
]

# Create an array of percentages
data = np.array([
    [24, 14, 7],   # Education
    [17, 12, 8],   # Agriculture
    [15, 10, 7],   # Poverty reduction
    [9, 5, 4],     # Reconstruction
    [7, 5, 5],     # Economic growth
    [3, 7, 6],     # Health
    [4, 6, 5],     # Job creation
    [5, 5, 4],     # Governance
    [4, 4, 6],     # Anti-corruption
    [4, 4, 4],     # Transport
    [3, 4, 4],     # Energy
    [2, 3, 4],     # Law & Justice
    [2, 3, 3],     # Basic infrastructure
    [2, 3, 3],     # Public sector reform
    [1, 3, 3]      # Public financial management
])

# Calculate total percentages
totals = data.sum(axis=1) # 1 means sum of each piorities

# Create figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Set up the positions
y_pos = np.arange(len(priorities))[::-1] # -1 mean last on top

# Colors for the three stacked segments
colors = ['darkblue', 'blue', 'lightblue']  # Dark to light blue for Most, 2nd, 3rd

# Plot horizontal stacked bars
bar_start = 15
left = np.ones(len(priorities)) * bar_start
for i in range(3):  # i=0: Most Important, i=1: 2nd Most, i=2: 3rd Most
    ax.barh(y_pos, data[:, i], left=left, color=colors[i], linewidth=0.1, height=0.6)
    left += data[:, i]

# Add priority names (first column) - right aligned
for i, priority in enumerate(priorities):
    ax.text(0, y_pos[i], priority, ha='right', va='center', fontsize=8, fontweight='normal')

# Add Total % values (second column) - center
for i, total in enumerate(totals):
    ax.text(8, y_pos[i], f"{int(total)}%", ha='center', va='center', fontsize=8)

# Add column headers (Third column) - center
ax.text(0, len(priorities), 'PRIORITY', ha='right', va='center', fontsize=7)
ax.text(12, len(priorities), 'TOTAL %', ha='right', va='center', fontsize=7)

# Add headers for the three stacked bars
ax.text(bar_start + 11, len(priorities), 'Most Important  |', ha='right', fontsize=8)
ax.text(bar_start + 25, len(priorities), '2nd Most Important  |', ha='right', fontsize=8)
ax.text(bar_start + 38, len(priorities), '3rd Most Important', ha='right', fontsize=8)

# Add percentage labels inside stacked bars
left_check = np.ones(len(priorities)) * bar_start
for i in range(3):
    for j in range(len(priorities)):
        val = data[j, i]
        if val > 0:
            # Position values
            pos = left_check[j]+1
            # White text on darker colors, black on lighter
            text_color = 'white' if i < 2 else 'black'
            ax.text(pos, y_pos[j], f"{int(val)}%", ha='center', va='center', 
                   color=text_color, fontsize=6)
    left_check += data[:, i]

# Remove axes spines and ticks
for spine in ax.spines.values():
    spine.set_visible(False)
ax.tick_params(axis='both', which='both', length=0, labelbottom=False, labelleft=False)

# Set limits - adjusted for better column widths
ax.set_xlim(0, 65)
ax.set_ylim(-1, len(priorities))

# Add note at bottom
ax.text(-2, -1.5, 'N=4,392. Respondents chose from a list. Top 15 shown.',
         fontsize=6, color='gray', ha='left')

plt.title('Top 15 development priorities, according to survey\n', loc='left')
plt.tight_layout()
plt.show()
```
